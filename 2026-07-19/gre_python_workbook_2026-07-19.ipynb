## 1. Text Completion

Text Completion tests whether the whole passage becomes logically, grammatically, and stylistically coherent.

### A reliable process

1. Read without looking at choices.
2. Mark structural clues: although, because, yet, therefore, despite, moreover.
3. Predict the blank in ordinary language.
4. Match the prediction to a choice.
5. Reread the completed sentence.

### Compare: vocabulary matching vs. logical completion

Suppose a sentence says:

> Although the proposal sounded radical, its actual recommendations were ___.

The word although signals contrast. A choice meaning “moderate” fits even if another option is a sophisticated synonym for “radical.” The test is not asking which choice is impressive; it is asking which choice completes the logic.

### Multi-blank consequence

Each blank has one correct choice, and there is no partial credit. Do not enumerate every possible combination. Solve the blank with the strongest clue first, then use it to constrain the others.

## 2. Sentence Equivalence

Sentence Equivalence contains one blank and six choices. Select **two** choices that both:

1. fit the sentence, and
2. make the completed sentences mean approximately the same thing.

### Compare: synonym hunting vs. sentence meaning

Two choices can be dictionary synonyms but fail the sentence. Conversely, correct choices need not be exact synonyms if the completed sentences convey equivalent meanings.

Predict first. Then evaluate every choice independently.

### Contrast clue example

> Far from being ___, the review eagerly praised even minor strengths.

Far from signals that the blank should mean reluctant or stingy with praise. A word meaning “negative” is too broad; the relation to **giving praise** matters.

## 3. Reading Comprehension

GRE Reading Comprehension is text-bound reasoning. Outside knowledge is irrelevant unless the passage supports it.

### Evidence discipline

For every answer, locate the sentence or relationship that proves it. A choice can be true in real life and still be wrong because it is not established by the passage.

### Compare: main point vs. interesting detail

A main-point answer must cover the passage's overall purpose. A choice mentioning only one study, mechanism, or example is usually too narrow. A choice making a stronger claim than the author is usually too broad.

### Compare: inference vs. speculation

A valid inference is strongly supported even if not stated word-for-word. A speculation is merely possible. Prefer the least extreme answer that must or very likely follows.

## 4. English usage and academic style

GRE Verbal does not contain a standalone grammar section, but Analytical Writing scoring considers clarity, sentence control, usage, and mechanics.

### Principle: structure carries meaning

Good academic prose is not prose with the largest vocabulary. It makes logical relationships and grammatical ownership clear.

### Compare: grammatical agreement

> The collection of reports **was** submitted.

The head noun is collection, not reports. Intervening plural nouns do not change the subject.

### Compare: dangling vs. attached modifier

> After reviewing the data, the conclusion seemed obvious.

Grammatically, the conclusion reviewed the data. Name the actor:

> After reviewing the data, the researchers found the conclusion obvious.

### Compare: concise vs. inflated

> Due to the fact that → Because  
> In the event that → If  
> Has the ability to → Can

Conciseness removes words without removing qualifications or logical relationships.

## 5. Quantitative Reasoning

GRE Quant emphasizes modeling and comparison more than lengthy computation.

### Quantitative Comparison choices

These are always:

A. Quantity A is greater  
B. Quantity B is greater  
C. The quantities are equal  
D. The relationship cannot be determined

### Compare: testing one value vs. testing the full domain

If x² = 9, testing x = 3 suggests x is positive—but x can also be -3. One valid example does not establish a relationship. Search for cases that produce different outcomes.

### Compare: percent changes vs. percentage-point changes

Increasing by 20% and then decreasing by 20% does not return to the starting value:

~~~text
100 × 1.20 × 0.80 = 96
~~~

The second percentage acts on a different base.

### Multiple-select discipline

Evaluate each option independently. The number of correct choices is not necessarily stated, and partial selection is incorrect.

### Numeric-entry discipline

Check units, requested form, and reasonableness. A probability must lie from 0 to 1; a count normally should be a nonnegative integer.
